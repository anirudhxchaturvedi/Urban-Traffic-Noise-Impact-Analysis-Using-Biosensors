```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'unsupervised_clustered_discomfort_levels_updated.csv'
data = pd.read_csv(file_path)

# Clean and preprocess the data
data['Sound Level (db)'] = data['Sound Level (db)'].str.extract(r'(\d+)').astype(float)
data = data.dropna(subset=['Subject', 'Sound Level (db)', 'Max Discomfort Level', 'Mean Discomfort Level'])

# Replace subject names with S1, S2, etc.
subject_mapping = {subject: f"S{i+1}" for i, subject in enumerate(sorted(data['Subject'].unique()))}
data['Subject'] = data['Subject'].map(subject_mapping)

# Ensure only numeric columns are aggregated
numeric_columns = data.select_dtypes(include='number').columns
averaged_data = data.groupby(['Subject', 'Sound Level (db)'], as_index=False)[numeric_columns].mean()

# Sort subjects based on numerical part of "S1", "S2", ...
averaged_data['Subject_Sort_Key'] = averaged_data['Subject'].str.extract(r'S(\d+)').astype(int)
averaged_data = averaged_data.sort_values(by=['Subject_Sort_Key', 'Sound Level (db)']).drop(columns='Subject_Sort_Key')

# Plot charts with averaged values for each sound level
sound_levels = averaged_data['Sound Level (db)'].unique()

for sound_level in sorted(sound_levels):
    subset = averaged_data[averaged_data['Sound Level (db)'] == sound_level]
    subjects = subset['Subject']
    max_discomfort = subset['Max Discomfort Level']
    mean_discomfort = subset['Mean Discomfort Level']
    
    # Plot the chart
    plt.figure(figsize=(10, 6))
    x = range(len(subjects))
    plt.bar(x, max_discomfort, width=0.4, label='Max Discomfort Level', alpha=0.7, align='center')
    plt.bar(x, mean_discomfort, width=0.4, label='Mean Discomfort Level', alpha=0.7, align='edge')
    plt.xticks(x, subjects, rotation=45, ha='right')
    plt.title(f'Discomfort Levels at {int(sound_level)} dB', fontsize=14)
    plt.xlabel('Subjects', fontsize=12)
    plt.ylabel('Discomfort Level (0-1)', fontsize=12)
    plt.ylim(0, 1)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Filter data for "Pink Noise" and group by sound level
if 'Source File_Discomfort' in data.columns:
    pink_noise_data = data[data['Source File_Discomfort'].str.contains('pink noise', case=False, na=False)]
    pink_noise_averaged = pink_noise_data.groupby(['Sound Level (db)'], as_index=False)[numeric_columns].mean()

    # Sort Pink Noise data by sound level
    pink_noise_averaged = pink_noise_averaged.sort_values(by='Sound Level (db)')

    # Plot chart for Pink Noise at different decibel levels
    plt.figure(figsize=(10, 6))
    x = range(len(pink_noise_averaged))
    plt.bar(x, pink_noise_averaged['Max Discomfort Level'], width=0.4, label='Max Discomfort Level', alpha=0.7, align='center')
    plt.bar(x, pink_noise_averaged['Mean Discomfort Level'], width=0.4, label='Mean Discomfort Level', alpha=0.7, align='edge')
    plt.xticks(x, pink_noise_averaged['Sound Level (db)'].astype(int), rotation=45, ha='right')
    plt.title('Averaged Discomfort Levels for Pink Noise at Different Decibel Levels', fontsize=14)
    plt.xlabel('Sound Level (dB)', fontsize=12)
    plt.ylabel('Discomfort Level (0-1)', fontsize=12)
    plt.ylim(0, 1)
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Source File_Discomfort' not found in the dataset.")
```

```python
import pandas as pd
import re

# Load the dataset
file_path = 'pink_noise_hr_filtered.csv'
data = pd.read_csv(file_path)

# Extract decibel levels (50, 60, 70) from the 'Source File_HR' column
def extract_decibel_level(source):
    match = re.search(r'\b(50|60|70)\b', str(source))
    return int(match.group(0)) if match else None

data['Decibel Level'] = data['Source File_HR'].apply(extract_decibel_level)

# Save the modified dataset to a new CSV file
output_file = 'updated_with_decibel_levels.csv'
data.to_csv(output_file, index=False)

print(f"Modified dataset with 'Decibel Level' column has been saved to {output_file}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'updated_with_decibel_levels.csv'
data = pd.read_csv(file_path)

# Ensure the 'Sound Level (db)' column is numeric, extracting values if necessary
if data['Sound Level (db)'].dtype != 'float64':
    data['Sound Level (db)'] = data['Sound Level (db)'].astype(str).str.extract(r'(\d+)').astype(float)

# Filter and clean the data
data = data.dropna(subset=['Sound Level (db)', 'Source File_HR', 'Max Discomfort Level', 'Mean Discomfort Level'])

# Get unique decibel levels
decibel_levels = sorted(data['Sound Level (db)'].unique())

# Generate charts for each decibel level
for level in decibel_levels:
    # Filter data for the current decibel level
    subset = data[data['Sound Level (db)'] == level].sort_values(by='Source File_HR')
    
    # Rename x-axis elements to S1, S2, S3, etc.
    subset['Custom_Label'] = [f"S{i+1}" for i in range(len(subset))]
    x_labels = subset['Custom_Label']
    max_discomfort = subset['Max Discomfort Level']
    mean_discomfort = subset['Mean Discomfort Level']
    
    # Plot the chart
    plt.figure(figsize=(12, 6))
    x = range(len(x_labels))
    bar_width = 0.4
    
    # Plot Max Discomfort Levels
    plt.bar(x, max_discomfort, width=bar_width, label='Max Discomfort Level', alpha=0.8)
    
    # Plot Mean Discomfort Levels (offset for grouping)
    plt.bar([pos + bar_width for pos in x], mean_discomfort, width=bar_width, label='Mean Discomfort Level', alpha=0.8)
    
    # Customize the chart
    plt.title(f'Discomfort Levels for {int(level)} dB', fontsize=16)
    plt.xlabel('Source File_HR (Renamed)', fontsize=14)
    plt.ylabel('Discomfort Level (0-1)', fontsize=14)
    plt.xticks([pos + bar_width / 2 for pos in x], x_labels, rotation=45, ha='right')
    plt.ylim(0, 1)
    plt.legend(fontsize=12)
    
    # Optimize layout
    plt.tight_layout()
    
    # Show the chart
    plt.show()
```
