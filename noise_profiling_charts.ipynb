```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your CSV file (update the path to your file)
file_path = 'C:/Users/HP/Downloads/noise profilling-20241119T185420Z-001/noise profilling updated/At 50db level/File 2/50dbfile2.csv'
df = pd.read_csv(file_path)

# Inspect the first few rows to identify the columns you want to use
print(df.head())

# Replace 'Column1' and 'Column2' with the names of the columns you want to plot
x_column = 'Time'  # Replace with the name of your X-axis column
y_column = 'OverallLevel'  # Replace with the name of your Y-axis column

# Plotting the chart
plt.figure(figsize=(12, 4))
plt.plot(df[x_column], df[y_column], color='blue', label='Overall Level vs Time')

x_min, x_max = df[x_column].min(), df[x_column].max()
x_ticks = np.linspace(x_min, x_max, num=10)  #evenly spaced divisions
plt.xticks(x_ticks, rotation=45)

y_min = 50
y_max = 80  # maximum value in your data
y_ticks = np.linspace(y_min, y_max, num=10)  # 10 equal parts
plt.yticks(y_ticks)

plt.title('Line Chart of Overall Level vs Time for Pink noise at 50Db', fontsize=16)  # Chart title
plt.xlabel('Time(sec)', fontsize=14)  # X-axis label
plt.ylabel('Overall Level(Db)', fontsize=14)  # Y-axis label
plt.legend()  # Ensure the legend is displayed
plt.grid(alpha=0.3)

output_path = 'C:/Users/HP/Downloads/noise profilling-20241119T185420Z-001/noise profilling updated/At 50db level/File 2/chart_output.png'
plt.savefig(output_path, dpi=300)
plt.show()
```
